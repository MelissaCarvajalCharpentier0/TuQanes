```python
# Preparación y SVM Clásica
import pandas as pd
import numpy as np
import optuna
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from imblearn.over_sampling import SMOTE

# ---------------------------------------------------------
# 1. CARGA DE DATOS
# ---------------------------------------------------------
# Asegúrate de subir tu archivo 'water_potability.csv' a Colab.
# Puedes hacerlo desde la barra lateral izquierda (ícono de carpeta).
try:
    df = pd.read_csv('/root/Downloads/water_potability.csv') # Updated path
    print("Dataset cargado exitosamente. Forma:", df.shape)
except FileNotFoundError:
    print("¡Ojo! Sube el archivo 'water_potability.csv' al entorno de Colab primero.")

# ---------------------------------------------------------
# 2. IMPUTACIÓN DE VALORES NULOS (Mediana por Clase)
# ---------------------------------------------------------
# Llenamos nulos de pH, Sulfato y Trihalometanos basados en si es potable (1) o no (0)
columnas_con_nulos = ['ph', 'Sulfate', 'Trihalomethanes'] # Changed 'pH' to 'ph'

for col in columnas_con_nulos:
    df[col] = df[col].fillna(df.groupby('Potability')[col].transform('median'))

print("\nValores nulos después de imputar:\n", df.isnull().sum())

# Separamos características (X) y objetivo (y)
X = df.drop('Potability', axis=1)
y = df['Potability']

# ---------------------------------------------------------
# 3. BALANCEO DE CLASES Y DIVISIÓN (80/20)
# ---------------------------------------------------------
# Usamos SMOTE para crear muestras sintéticas de la clase minoritaria (Potable)
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X, y)

print(f"\nDistribución de clases ANTES de SMOTE:\n{y.value_counts()}")
print(f"Distribución de clases DESPUÉS de SMOTE:\n{y_bal.value_counts()}")

# División estratificada 80% entrenamiento, 20% prueba
X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size=0.20, random_state=42, stratify=y_bal)

# ---------------------------------------------------------
# 4. ESTANDARIZACIÓN (Fundamental para SVM y Cuántica)
# ---------------------------------------------------------
# El StandardScaler se ajusta SOLO con los datos de entrenamiento para evitar fuga de datos (data leakage)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------------------------------------------------------
# 5. OPTIMIZACIÓN DE HIPERPARÁMETROS CON OPTUNA (Nivel Pro)
# ---------------------------------------------------------
def objective(trial):
    # Optuna buscará el mejor C y gamma en estos rangos continuos
    c_param = trial.suggest_float('C', 0.1, 100.0, log=True)
    gamma_param = trial.suggest_categorical('gamma', ['scale', 'auto', 0.01, 0.1, 1])

    # Creamos el modelo clásico con RBF Kernel
    svc = SVC(kernel='rbf', C=c_param, gamma=gamma_param, random_state=42)

    # Evaluamos usando validación cruzada (5 folds) optimizando el F1-Score
    score = cross_val_score(svc, X_train_scaled, y_train, cv=5, scoring='f1').mean()
    return score

print("\n--- Iniciando Optuna para buscar hiperparámetros óptimos ---")
study = optuna.create_study(direction='maximize')
# Limitamos a 20 intentos (trials) para no tardar mucho, pero puedes subirlo
study.optimize(objective, n_trials=20)

print(f"\nMejores hiperparámetros encontrados: {study.best_params}")
print(f"Mejor F1-Score en Validación Cruzada: {study.best_value:.4f}")

# ---------------------------------------------------------
# 6. ENTRENAMIENTO FINAL Y EVALUACIÓN LÍNEA BASE CLÁSICA
# ---------------------------------------------------------
# Entrenamos el modelo final con los mejores parámetros
best_svc = SVC(kernel='rbf', **study.best_params, random_state=42)
best_svc.fit(X_train_scaled, y_train)

# Predicción en el conjunto de PRUEBA (datos nunca antes vistos)
y_pred = best_svc.predict(X_test_scaled)

print("\n================ REPORTE SVM CLÁSICA ================")
print(confusion_matrix(y_test, y_pred))
print("\n", classification_report(y_test, y_pred))

# ---------------------------------------------------------
# 7. EXTRACCIÓN DEL SUBCONJUNTO PARA EL KERNEL CUÁNTICO
# ---------------------------------------------------------
# Extraemos exactamente 64 muestras estratificadas (32 potables, 32 no potables)
# de nuestro conjunto de prueba escalado.
splitter = StratifiedShuffleSplit(n_splits=1, train_size=64, random_state=42)

for sub_idx, _ in splitter.split(X_test_scaled, y_test):
    X_quantum_subset = X_test_scaled[sub_idx]
    y_quantum_subset = y_test.iloc[sub_idx].values # Convertimos a array numpy

print("\n================ SUBCONJUNTO CUÁNTICO PREPARADO ================")
print(f"Tamaño X cuántico: {X_quantum_subset.shape}")
print(f"Tamaño y cuántico: {y_quantum_subset.shape}")
print(f"Balance de clases cuántico: {np.bincount(y_quantum_subset)}")
print("¡Todo listo para la Parte 2 (Circuitos Cuánticos)!")
```
